# Merging The Country Codes to their Repsected Countries.

```python
plastic_waste = plastic_df.merge(code_df[['Country', 'alpha-3']], on='Country', how='left')

# Rename column for clarity (optional)
plastic_waste.rename(columns={'alpha-3': 'Country_Code'}, inplace=True)

# Display the updated DataFrame
plastic_waste.head()
```

# Since we have missing data based on the code , creatd a dict. of coutries codes that didn't contain their countries code and add it to it.

```python
missing_codes = plastic_waste [plastic_waste ['Country_Code'].isna()]

# Display the countries with missing codes
print(missing_codes[['Country']])
```

```python
# Dictionary mapping missing countries to their correct ISO 3-letter codes
missing_country_codes = {
    "United States": "USA",
    "Russia": "RUS",
    "United Kingdom": "GBR",
    "Turkey": "TUR",
    "South Korea": "KOR",
    "Vietnam": "VNM",
    "Netherlands": "NLD",
    "Czech Republic": "CZE",
    "Bolivia": "BOL",
    "Laos": "LAO",
    "Syria": "SYR",
    "Tanzania": "TZA",
    "Cape Verde": "CPV",
    "Ivory Coast": "CIV",
    "Republic of Congo": "COG",
    "Democratic Republic of Congo": "COD",
    "Moldova": "MDA",
    "Kosovo": "XKX",
    "Brunei": "BRN",
    "East Timor": "TLS",
    "Micronesia": "FSM"
}

# Only update rows where 'Country_Code' is NaN
plastic_waste['Country_Code'] = plastic_waste.apply(
    lambda row: missing_country_codes[row['Country']] if pd.isna(row['Country_Code']) and row['Country'] in missing_country_codes else row['Country_Code'],
    axis=1
)


plastic_waste

```
